```python
a = pd.read_csv("C/LIBYA.csv")
```

```python
def frequency_table(x):
    return pd.crosstab(index=x,  columns="count")
```

```python
ctabs = {}
for column in a:
    ctabs[column]=frequency_table(a[column])
```
